### Transform Location to Lat/Lon

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='myapplication')
```

```python
geolocator.geocode("Alhambra, CA, USA")
```

```python
# create a lat/lon dict for each location
locations = df_ba['Location'].unique()

def get_lat_lon(loc):
    """Get lon/lat coordinates for the location
    
    USA must be added because place names can be ambiguous
    """
    loc += ", USA"
    location = geolocator.geocode(loc)
    return [location.raw['lat'], location.raw['lon']]

latlons = {loc: get_lat_lon(loc) for loc in locations}
```

```python
df_ba['lat'] = df_ba['Location'].apply(lambda x: float(latlons[x][0]))
```

```python
df_ba['lon'] = df_ba['Location'].apply(lambda x: float(latlons[x][1]))
```

```python
df_ba[['Location', 'lat', 'lon']]
```

### Transform salaries to numbers

```python
import re

def get_min_max_salary(salary_str):
    """Use regex to extract the minimum and maximum salaries from the Salary Estimate field
    """
    m = re.findall("\$([0-9]+)K*", salary_str)
    min_s, max_s = int(m[0]) * 1000, int(m[1]) * 1000
    return min_s, max_s

assert get_min_max_salary('$56K-$102K (Glassdoor est.)') == (56000, 102000)
```

```python
df_ba['salary_min'] = df_ba['Salary Estimate'].apply(lambda x: get_min_max_salary(x)[0])
df_ba['salary_max'] = df_ba['Salary Estimate'].apply(lambda x: get_min_max_salary(x)[1])
```

```python
df_ba[['salary_min', 'salary_max']].describe()
```

# Modelling

# Q1: What are the geographical hotspot for finding a data job?
# Heatmap of jobs by location

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
counts = np.zeros((16, 24))

# linear bins from min to max
lat_min = df_ba['lat'].min()
lat_max = df_ba['lat'].max()+0.01
lon_min = df_ba['lon'].min()
lon_max = df_ba['lon'].max()+0.01

def get_bin(lat, lon):
    """2D bin calculations from lat/lon coordinates
    
    Number is bins is hardcoded
    """
    lat -= lat_min
    lat /= lat_max - lat_min
    lat *= 16
    
    lon -= lon_min
    lon /= lon_max - lon_min
    lon *= 24
    
    return int(lat), int(lon)

for lat, lon in zip(df_ba['lat'], df_ba['lon']):
    lat_idx, lon_idx = get_bin(lat, lon)
    counts[lat_idx, lon_idx] += 1
```

```python
plt.figure(figsize=(24*0.6,16*0.6))
plt.imshow(np.flip(counts, axis=0), cmap='hot', interpolation='nearest')
plt.show()
```

It's obvious there are a few hotspots, however the data is hard to interpret

```python
# create a dictionary from locations to prevalance for each locations
count_dict = {k: v for k, v in zip(df_ba['Location'].value_counts().index, df_ba['Location'].value_counts().values)}
```

```python
df_ba['locCount'] = df_ba['Location'].apply(lambda x: count_dict[x])
```

```python
# visualise prevalence of jobs on their lat/lon coordinate
plt.figure(figsize=(20,10))
plt.scatter(df_ba['lon'], df_ba['lat'], s=0.1+df_ba['locCount']/df_ba['locCount'].max())
```

Using the raw lat/lon coords, West coast CA and East coast NY are clearly visible as hotspots.
